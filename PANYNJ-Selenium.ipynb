## Function for scraping
I created a function for scraping modeling it on code from a template by Soma. Some tweaks: I knew I would get bad/invalid results. So I needed try/except clauses. The scrape was long, something like an hour as the browser clicked through 598 entries.

```python
driver = webdriver.Chrome()
def process_origins(row):
    
    # Visit the search page
    driver.get('https://www.imdb.com')
    
    # Fill out the form and submit it
    try:
        imdb_input = driver.find_element(By.XPATH, '//*[@id="suggestion-search"]')
        imdb_input.send_keys(row['name'])
        driver.find_element(By.XPATH, '//*[@id="suggestion-search-button"]').click()
        driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/div/div[1]/section[2]/div[2]/ul/li[1]/div[2]/div[1]/a').click()
        country_of_origin = driver.find_element(By.CSS_SELECTOR, 'li[data-testid="title-details-origin"]')
        print(country_of_origin.text)
        print("------------")
        return pd.Series({
            'country_of_origin': country_of_origin.text
            })
    except:   
         
        return pd.Series({
            'country_of_origin': "No result"
            })

result_df = df.apply(process_origins, axis=1).join(df)
  
#result_df = df.apply(process_origins, axis=1).join(df)
    #origin_df = df.apply(process_origins, axis=1).join(df)
```
